İstifadəçinin daxil etdiyi location-dan daxil etdiyi datanı oxudan funksiyanı yazın

Input : (location, dataname )

Output : return data from that location

```python
dataname='CARS'
extension='.xlsx'
```

```python
def function4(location,dataname,filetype):
    if filetype=='excel':
        fullpath=location+dataname+'.xlsx'
        return pd.read_excel(fullpath)
    else:
        fullpath=location+dataname+'.csv'
        return pd.read_csv(fullpath)
```

#### 7

İstifadəçinin daxil etdiyi 2 string, daxil edilmiş datada sütun adını dəyişdirməyə kömək etsin. Yəni stringlərdən biri köhnə column, digəri isə yeni column üçündür

Input : (data, oldcolumn, newcolumn )

Output : return replaced column name(s)

```python
def function7(dataset,oldcolumn,newcolumn):
    data.rename(columns={oldcolumn:newcolumn},inplace=True)
    return dataset
```

```python
function7(data,'Gender','GGG')
```
