```python
#9) add column for furnished rows as good, semi-furnished as standard, the others as bad condition
data.loc[data['furnishing_status']=='furnished', 'furnishing_condition']='Good'
data.loc[data['furnishing_status']=='semi-furnished', 'furnishing_condition']='Standart'
data.loc[data['furnishing_status']!='furnished', 'furnishing_condition']='Bad'
```

```python
#10) add column for if price per area is lower than its median number, set value as 'Cheap', Else set value as 'Expensive'
data['price_per_area_status']=np.where(data['price_per_area']< data['price_per_area'].median(),'Cheap','Expensive')
```
